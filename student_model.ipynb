```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
student_df = pd.read_csv("student_data.csv")
student_df.head()
```

```python
student_df.info()
```

```python
 student_data = student_df.drop(["Interest","Student ID"],axis = 1)
student_data.head()
```

```python
    student_data.head(1000)
```

```python
student_data['Stream'] = student_data['Recommended Stream'].map({'Arts': -1, 'Commerce': 0, 'Science': 1})
student_data.head()
```

```python
df = student_data.drop(['Recommended Stream','Career Interest'],axis = 1)
df.head()
```

```python
df.describe()
```

```python
x = df.drop(["Stream"],axis = 1)
y = df["Stream"]
```

```python
from sklearn.model_selection import train_test_split

def classify(model, x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    # train the model
    model.fit(x_train, y_train)
    print("Accuracy:", model.score(x_test, y_test) * 100)
    return model
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
model = LogisticRegression(max_iter=1000, solver='lbfgs')  # Adjust solver if needed

# Train and evaluate using the classify function
LinearReg_acc = classify(model, x, y)
print("Logistic Regression Accuracy:", LinearReg_acc)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=1)
SVM_acc = classify(model, x, y)
print("SVM Accuracy:", SVM_acc)
```
