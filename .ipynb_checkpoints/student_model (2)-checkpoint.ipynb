```python
x = onehot_encoded_df.drop(["Recommended Stream_Arts","Recommended Stream_Commerce","Recommended Stream_Science"],axis = 1)
y = onehot_encoded_df[["Recommended Stream_Arts","Recommended Stream_Commerce","Recommended Stream_Science"]]
```

```python
from imblearn.over_sampling import SMOTE

# Initialize the SMOTE object with the desired parameters
oversample = SMOTE(k_neighbors=5)  # k_neighbors=5 is a good default value

# Apply SMOTE to the features (x) and target labels (y)
x_resampled, y_resampled = oversample.fit_resample(x, y)

# Verify the class distribution after resampling
print("Before SMOTE:")
print(y.value_counts())  # Original class distribution
print("\nAfter SMOTE:")
print(y_resampled.value_counts())
```

```python
from sklearn.model_selection import train_test_split

def classify(model, x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    # train the model
    model.fit(x_train, y_train)
    print("Accuracy:", model.score(x_test, y_test) * 100)
    return model.score(x_test, y_test) * 100
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Scale the features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Initialize the Logistic Regression model
model = LogisticRegression(max_iter=1000, solver='lbfgs')  # Adjust solver if needed

# Train and evaluate using the classify function
LinearReg_acc = classify(model, x_scaled, y)
print("Logistic Regression Accuracy:", LinearReg_acc)
```

```python
import joblib
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

# Scale the features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Initialize the SVC model
model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=1)

# Train and evaluate using the classify function
SVM_acc = classify(model, x_scaled, y)
print("SVM Accuracy:", SVM_acc)

# Save the trained model using joblib
joblib.dump(model, 'svm_model.pkl')
joblib.dump(scaler, 'scaler.pkl') 
```
